```python
import pandas as pd

# Load the dataset
url_red = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
url_white = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv"
df_red = pd.read_csv(url_red, sep=';')
df_white = pd.read_csv(url_white, sep=';')

# Add a type column to differentiate between red and white wines
df_red['type'] = 'red'
df_white['type'] = 'white'

# Combine the datasets
df = pd.concat([df_red, df_white], ignore_index=True)

# Display the first few rows of the combined dataset
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of wine types
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='type', palette='viridis')
plt.title('Distribution of Red vs. White Wines')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()

# Distribution of quality ratings
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='quality', palette='viridis')
plt.title('Distribution of Wine Quality Ratings')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()

# Boxplot of quality ratings by wine type
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='quality', y='type', palette='viridis')
plt.title('Quality Ratings by Wine Type')
plt.xlabel('Quality')
plt.ylabel('Type')
plt.show()
```

```python
df['type'] = df['type'].astype('category').cat.codes
correlation_matrix = df.corr()
```

```python
# Correlation heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap of Wine Features')
plt.show()

# Alcohol content distribution
plt.figure(figsize=(12, 8))
sns.histplot(data=df, x='alcohol', hue='type', element='step', palette='viridis', kde=True)
plt.title('Distribution of Alcohol Content by Wine Type')
plt.xlabel('Alcohol Content')
plt.ylabel('Count')
plt.show()

# Sulphates vs. Quality
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='sulphates', y='quality', hue='type', palette='viridis')
plt.title('Sulphates vs. Quality by Wine Type')
plt.xlabel('Sulphates')
plt.ylabel('Quality')
plt.show()
```
